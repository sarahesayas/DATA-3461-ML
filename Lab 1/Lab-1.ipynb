2. Create functions model_loop(x,w,b) and model_np(x,w,b) that compute y=wx+b
, the first using explicit loops, the second computing using numpy and no loops. Illustrate that the two functions create identical output. Use numpy version for the remainder of the lab.

```python
def model_loop(x, w, b):
    y = np.zeros_like(x) 
    for i in range(len(x)):
        y[i] = w * x[i] + b
    return y

y_loop = model_loop(x, w, b)
```

```python
def model_np(x, w, b):
    return w * x + b

y_np = model_np(x, w, b)
```

```python
are_close = np.allclose(y_loop, y_np)
print(f"Outputs are identical: {are_close}")
```

4. Create a function `cost(x,y,w,b)` that computes the mean square error between model prediction and data. Use numpy expressions and do not do any explicit loops in python. Make 2 plots illusrating the parabolic shape of the cost function:
   * Plot the cost vs $w$ for the dataset, holding $b$ constant. 
   * Plot the cost vs $b$ for the dataset, holding $w$ constant. 


```python
def cost(x, y, w, b):
    predictions = w * x + b
    mse = np.mean((y - predictions) ** 2)
    return mse

# Define ranges for w and b to explore
w_values = np.linspace(0, 6, 100)
b_values = np.linspace(0, 15, 100)

# Compute cost for different values of w, holding b constant
b_fixed = 7.5
cost_vs_w = [cost(x, y, w, b_fixed) for w in w_values]

# Compute cost for different values of b, holding w constant
w_fixed = 3
cost_vs_b = [cost(x, y, w_fixed, b) for b in b_values]

# Plot cost vs. w
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(w_values, cost_vs_w, label='Cost vs. w')
plt.title('Cost vs. w')
plt.xlabel('w')
plt.ylabel('Cost')
plt.grid(True)

# Plot cost vs. b
plt.subplot(1, 2, 2)
plt.plot(b_values, cost_vs_b, label='Cost vs. b', color='orange')
plt.title('Cost vs. b')
plt.xlabel('b')
plt.ylabel('Cost')
plt.grid(True)

plt.legend()
plt.tight_layout()
plt.show()
```

5. Create a function `dj_dw, dj_db = gradient(x, y, w, b)` that computes the gradient wrt $w$ and $b$ on a dataset. Use numpy expressions and do not do any explicit loops in python. Make 2 plots that illustrates that the gradient is 0 at the optimal values of $w$ and $b$ by:
   * Ploting the gradient vs $w$ for the dataset, holding $b$ constant. 
   * Ploting the gradient vs $b$ for the dataset, holding $w$ constant. 

```python
# Function to compute gradients
def gradient(x, y, w, b):
    n = len(x)
    predictions = w * x + b
    dj_dw = -2 / n * np.sum((y - predictions) * x)
    dj_db = -2 / n * np.sum(y - predictions)
    return dj_dw, dj_db

# Define ranges for w and b to explore
w_values = np.linspace(0, 6, 100)
b_values = np.linspace(0, 15, 100)

# Compute gradients for different values of w, holding b constant
b_fixed = 7.5
gradients_w = [gradient(x, y, w, b_fixed)[0] for w in w_values]

# Compute gradients for different values of b, holding w constant
w_fixed = 3
gradients_b = [gradient(x, y, w_fixed, b)[1] for b in b_values]

# Plot gradient vs. w
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(w_values, gradients_w, label='Gradient vs. w')
plt.title('Gradient vs. w')
plt.xlabel('w')
plt.ylabel('Gradient')
plt.grid(True)

# Plot gradient vs. b
plt.subplot(1, 2, 2)
plt.plot(b_values, gradients_b, label='Gradient vs. b', color='orange')
plt.title('Gradient vs. b')
plt.xlabel('b')
plt.ylabel('Gradient')
plt.grid(True)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import copy
import math

# Gradient Descent function
def gradient_descent(x, y, w_in, b_in, alpha, num_iters, cost, gradient):
    w = copy.deepcopy(w_in)
    b = b_in
    
    J_history = []
    p_history = []

    for i in range(num_iters):
        dj_dw, dj_db = gradient(x, y, w, b)
        
        b = b - alpha * dj_db
        w = w - alpha * dj_dw
        
        if i < 100000:
            J_history.append(cost(x, y, w, b))
            p_history.append([w, b])
        
        if i % math.ceil(num_iters / 10) == 0:
            print(f"Iteration {i:4}: Cost {J_history[-1]:0.2e} ",
                  f"dj_dw: {dj_dw: 0.3e}, dj_db: {dj_db: 0.3e}  ",
                  f"w: {w: 0.3e}, b:{b: 0.5e}")
    
    return w, b, J_history, p_history

# Create the dataset
np.random.seed(42)
n_samples = 1000
w_true = 3
b_true = 7.5
noise_std = 40

x = np.random.rand(n_samples) * 100  
y = w_true * x + b_true + np.random.randn(n_samples) * noise_std  

# Normalize data
x_mean = np.mean(x)
x_std = np.std(x)
x_norm = (x - x_mean) / x_std

# Initialize parameters
w_init = 0
b_init = 0
alpha = 0.0001  # Reduced learning rate
num_iters = 1000  # Number of iterations

# Run gradient descent
w_final, b_final, J_history, p_history = gradient_descent(x_norm, y, w_init, b_init, alpha, num_iters, cost, gradient)

# Print final results
print(f"Final parameters: w = {w_final:.3f}, b = {b_final:.3f}")

# Plot cost function history
plt.figure(figsize=(12, 6))
plt.plot(J_history, label='Cost over iterations')
plt.xlabel('Number of iterations')
plt.ylabel('Cost')
plt.title('Cost Function History')
plt.grid(True)
plt.legend()
plt.show()

# Plot final line against the data
plt.figure(figsize=(12, 6))
plt.scatter(x, y, alpha=0.5, label='Data points')
plt.plot(np.sort(x), w_final * np.sort(x) + b_final, color='red', label='Fitted line', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data and Fitted Line')
plt.legend()
plt.grid(True)
plt.show()
```
